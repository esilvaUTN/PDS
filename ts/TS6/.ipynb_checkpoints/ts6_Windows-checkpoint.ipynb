### Configuración e inicio de la simulación

### Cálculo de estimador de Frecuencia.

```python
for snr in SNR_db:
    SNR = 10**(snr/10)      #SNR en veces
    Pn = ((a1**2)/2)/SNR    #Potencia de ruido
    
    for i in range(R):
        W = w1[i]
        s[:, i] = a1*np.sin(2*np.pi*W*k)                  #Generación de señal senoidal
        n[:, i] = np.random.normal(0, np.sqrt(Pn), N)     #Ruido aleatorio analógico
    
    x = s + n   #Señal con ruido
    
    f = np.fft.fftfreq(len(x), 1/fs)    #Vector de frecuencias
    ft_X = np.fft.fft(x, axis=0)/N      #Espectro
    
    index = np.argmax(np.abs(ft_X[:fs//2, :]), axis=0)  #Estimador de fase
    w = f[index]
    plt.hist(w, bins=20, alpha=0.8, label=f'{snr} dB')
    plt.title('Histograma del Estimador de Frecuencia')
    plt.xlabel('Frecuencia')
    plt.ylabel('N')
    plt.legend()

    print(f"Media para SNR = {snr}dB: " + str(np.mean(w)))
    print(f"Sesgo para SNR = {snr}dB: " + str(np.mean(w1-w)))
    print(f"Varianza para SNR = {snr}dB: " + str(np.var(w)) + "\n")
```

### Se observa que tanto en módulo como en fase, a mayor SNR, el histograma se encuentra menos disperso, es decir, que posee menor varianza.
